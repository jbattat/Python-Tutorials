```python
import matplotlib.pyplot as plt
import numpy as np 

# Sample randomly to give equal number of points per area of sphere
# See: https://mathworld.wolfram.com/SpherePointPicking.html
n=1000
u = np.random.uniform(0, 1, n)
v = np.random.uniform(0, 1, n)

phi = 2*np.pi*u - np.pi          # longitude (from -pi to pi)
theta = np.arccos(2*v-1)-np.pi/2 # latitude (from -pi/2 to pi/2)

#print(f"phi:   min, max = {np.min(phi)}, {np.max(phi)}")
#print(f"theta: min, max = {np.min(theta)}, {np.max(theta)}")

# Make figure
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(projection='hammer' ) # or mollweide, aitoff, or lambert. [example](http://matplotlib.org/examples/pylab_examples/geo_demo.html) 
ax.scatter(phi, theta, marker='.', color='r')
ax.grid(True)
ax.set_xlabel(r'$\phi$')
ax.set_ylabel(r'$\theta$')
ax.set_title("Hammer-Aitoff projection (equal-area)");
```
